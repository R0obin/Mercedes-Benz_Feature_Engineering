```python
def one_hot_encode(data,variable,X_labels):
  for label in X_labels:
    data[variable +'_' + label]= np.where(data[variable]==label,1,0)

df = pd.read_csv('/content/test.csv',usecols=['X1','X2','X3','X4','X5','X6'])
```

```python
top_ten_1 = [feature for feature in df.X1.value_counts().sort_values(ascending=False).head(10).index]
top_ten_1
```

```python
one_hot_encode(df,'X1',top_ten_1)
df.head(20)
```
